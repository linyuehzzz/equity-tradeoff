```python
import geopandas as gpd
import pandas as pd
import ast
from shapely.geometry import Point
from shapely.wkt import loads
```

Attribute values

```python
chicago = gpd.read_file('data/chicago_tract_geo.shp')
chicago_tbl1 = pd.read_csv('data/nhgis0014_ds262_20225_tract.csv', encoding='latin-1')
chicago_tbl2 = pd.read_csv('data/nhgis0014_ds263_20225_tract.csv', encoding='latin-1')
chicago_all = chicago.merge(chicago_tbl1, on='GISJOIN')
chicago_all = chicago_all.merge(chicago_tbl2, on='GISJOIN')

# age: % age < 18 or > 64
chicago["AGE"] = (sum([chicago_all["AQM4E0%s" % '{number:0{width}d}'.format(width=2, number=i)] for i in range(3, 7)]) + \
    sum([chicago_all["AQM4E0%s" % '{number:0{width}d}'.format(width=2, number=i)] for i in range(20, 26)]) + \
    sum([chicago_all["AQM4E0%s" % '{number:0{width}d}'.format(width=2, number=i)] for i in range(27, 31)]) + \
    sum([chicago_all["AQM4E0%s" % '{number:0{width}d}'.format(width=2, number=i)] for i in range(44, 50)])) / chicago_all["AQM4E001"]

# race: % non-white
chicago["RACE"] = (sum([chicago_all["AQNGE0%s" % '{number:0{width}d}'.format(width=2, number=i)] for i in range(3, 9)])) / chicago_all["AQNGE001"]

# ethnicity: % hispanic
chicago["ETH"] = chicago_all["AQNPE003"] / chicago_all["AQNPE001"]

# education: % with high school diploma or below
chicago["EDU"] = (sum([chicago_all["AQPJE0%s" % '{number:0{width}d}'.format(width=2, number=i)] for i in range(3, 12)]) + \
                  sum([chicago_all["AQPJE0%s" % '{number:0{width}d}'.format(width=2, number=i)] for i in range(20, 29)])) / chicago_all["AQPJE001"]

# language: % speaker language other than English
chicago["LAN"] = (sum([chicago_all["AQPXE0%s" % '{number:0{width}d}'.format(width=2, number=i)] for i in range(3, 15)])) / chicago_all["AQPXE001"]

# household income: % less than 45000
chicago["INC"] = (sum([chicago_all["AQP5E0%s" % '{number:0{width}d}'.format(width=2, number=i)] for i in range(2, 10)])) / chicago_all["AQP5E001"]

# computer access: % no computer
chicago["COM"] = chicago_all["AQWFE007"] / chicago_all["AQWFE001"]

# internet access: % no internet
chicago["INT"] = chicago_all["AQWGE008"] / chicago_all["AQWGE001"]

# disability: % with disability
chicago["DIS"] = (sum([chicago_all["AQ7BE0%s" % '{number:0{width}d}'.format(width=2, number=i)] for i in range(4, 21, 3)]) + \
                  sum([chicago_all["AQ7BE0%s" % '{number:0{width}d}'.format(width=2, number=i)] for i in range(23, 40, 3)])) / chicago_all["AQ7BE001"]

chicago
```

Tech sites

```python
# Read the CSV file
tech_sites = pd.read_csv('data/Public_Technology_Resources_20240705.csv', encoding='latin-1')

# Extract coordinates and convert to separate 'lat' and 'lon' columns
tech_sites['coord'] = tech_sites['LOCATION'].str.split('\n').str[2].apply(ast.literal_eval)
tech_sites['lat'] = tech_sites['coord'].apply(lambda x: x[0])
tech_sites['lon'] = tech_sites['coord'].apply(lambda x: x[1])

# Convert to GeoDataFrame by creating a geometry column with Point objects
tech_sites['geometry'] = tech_sites.apply(lambda row: Point(row['lon'], row['lat']), axis=1)
tech_sites_gdf = gpd.GeoDataFrame(tech_sites, geometry='geometry', crs=4326).to_crs(chicago.crs)

# Drop the intermediate 'coord', 'lat', and 'lon' columns if not needed
tech_sites_gdf = tech_sites_gdf.drop(columns=['coord', 'lat', 'lon'])
tech_sites_gdf.head()
```

```python
tech_sites_gdf['TYPE'].unique()
```

```python
# Perform spatial join
tech_sites_joined = tech_sites_gdf.sjoin(chicago[['GISJOIN', 'geometry']], how='left')

# Count occurrences of OBJECTID by GISJOIN
count = tech_sites_joined.groupby('GISJOIN').size().reset_index(name='TECH_CNT')

# Merge count with chicago GeoDataFrame
chicago = chicago.merge(count, on='GISJOIN', how='left')

# Fill NaN values with 0 and convert to integer
chicago['TECH_CNT'] = chicago['TECH_CNT'].fillna(0).astype(int)
chicago
```

Max capacity

```python
# libraries
libraries = pd.read_csv('data/Libraries_-_2022_Visitors_by_Location_20240706.csv', encoding='latin-1')
libraries['geometry'] = libraries['Location'].apply(loads)
libraries_gdf = gpd.GeoDataFrame(libraries, geometry='geometry', crs=4326).to_crs(chicago.crs)

# schools
schools = pd.read_csv('data/CPS_School_Locations_SY1415_20240706.csv', encoding='latin-1')
schools['geometry'] = schools['the_geom'].apply(loads)
schools_gdf = gpd.GeoDataFrame(schools, geometry='geometry', crs=4326).to_crs(chicago.crs)

# community centers
community_centers = pd.read_csv('data/Community_Service_Centers_20240706.csv', encoding='latin-1')
community_centers['coord'] = community_centers['LOCATION'].str.split('\n').str[2].apply(ast.literal_eval)
community_centers['lat'] = community_centers['coord'].apply(lambda x: x[0])
community_centers['lon'] = community_centers['coord'].apply(lambda x: x[1])
community_centers['geometry'] = community_centers.apply(lambda row: Point(row['lon'], row['lat']), axis=1)
community_centers_gdf = gpd.GeoDataFrame(community_centers, geometry='geometry', crs=4326).to_crs(chicago.crs)
community_centers_gdf = community_centers_gdf.drop(columns=['coord', 'lat', 'lon'])

# senior centers
senior_centers = pd.read_csv('data/Senior_Centers_20240706.csv', encoding='latin-1')
senior_centers['coord'] = senior_centers['LOCATION'].str.split('\n').str[2].apply(ast.literal_eval)
senior_centers['lat'] = senior_centers['coord'].apply(lambda x: x[0])
senior_centers['lon'] = senior_centers['coord'].apply(lambda x: x[1])
senior_centers['geometry'] = senior_centers.apply(lambda row: Point(row['lon'], row['lat']), axis=1)
senior_centers_gdf = gpd.GeoDataFrame(senior_centers, geometry='geometry', crs=4326).to_crs(chicago.crs)
senior_centers_gdf = senior_centers_gdf.drop(columns=['coord', 'lat', 'lon'])

# workforce centers
workforce_centers = pd.read_csv('data/Workforce_Centers_20240706.csv', encoding='latin-1')
workforce_centers['coord'] = workforce_centers['LOCATION'].str.split('\n').str[2].apply(ast.literal_eval)
workforce_centers['lat'] = workforce_centers['coord'].apply(lambda x: x[0])
workforce_centers['lon'] = workforce_centers['coord'].apply(lambda x: x[1])
workforce_centers['geometry'] = workforce_centers.apply(lambda row: Point(row['lon'], row['lat']), axis=1)
workforce_centers_gdf = gpd.GeoDataFrame(workforce_centers, geometry='geometry', crs=4326).to_crs(chicago.crs)
workforce_centers_gdf = workforce_centers_gdf.drop(columns=['coord', 'lat', 'lon'])

# family and support services
support_service = pd.read_csv('data/Family_and_Support_Services_Delegate_Agencies.csv', encoding='latin-1')
support_service['geometry'] = support_service.apply(lambda row: Point(row['Longitude'], row['Latitude']), axis=1)
support_service_gdf = gpd.GeoDataFrame(support_service, geometry='geometry', crs=4326).to_crs(chicago.crs)
```

```python
# libraries
libraries_joined = libraries_gdf.sjoin(chicago[['GISJOIN', 'geometry']], how='left')
count = libraries_joined.groupby('GISJOIN').size().reset_index(name='LIB_CNT')
chicago = chicago.merge(count, on='GISJOIN', how='left')
chicago['LIB_CNT'] = chicago['LIB_CNT'].fillna(0).astype(int)

# schools
schools_joined = schools_gdf.sjoin(chicago[['GISJOIN', 'geometry']], how='left')
count = schools_joined.groupby('GISJOIN').size().reset_index(name='SCH_CNT')
chicago = chicago.merge(count, on='GISJOIN', how='left')
chicago['SCH_CNT'] = chicago['SCH_CNT'].fillna(0).astype(int)

# community centers
community_centers_joined = community_centers_gdf.sjoin(chicago[['GISJOIN', 'geometry']], how='left')
count = community_centers_joined.groupby('GISJOIN').size().reset_index(name='CC_CNT')
chicago = chicago.merge(count, on='GISJOIN', how='left')
chicago['CC_CNT'] = chicago['CC_CNT'].fillna(0).astype(int)

# senior centers
senior_centers_joined = senior_centers_gdf.sjoin(chicago[['GISJOIN', 'geometry']], how='left')
count = senior_centers_joined.groupby('GISJOIN').size().reset_index(name='SC_CNT')
chicago = chicago.merge(count, on='GISJOIN', how='left')
chicago['SC_CNT'] = chicago['SC_CNT'].fillna(0).astype(int)

# workforce centers
workforce_centers_joined = workforce_centers_gdf.sjoin(chicago[['GISJOIN', 'geometry']], how='left')
count = workforce_centers_joined.groupby('GISJOIN').size().reset_index(name='WC_CNT')
chicago = chicago.merge(count, on='GISJOIN', how='left')
chicago['WC_CNT'] = chicago['WC_CNT'].fillna(0).astype(int)

# family and support services
support_service_joined = support_service_gdf.sjoin(chicago[['GISJOIN', 'geometry']], how='left')
count = support_service_joined.groupby('GISJOIN').size().reset_index(name='SS_CNT')
chicago = chicago.merge(count, on='GISJOIN', how='left')
chicago['SS_CNT'] = chicago['SS_CNT'].fillna(0).astype(int)

chicago['MAX_CAP'] = chicago['LIB_CNT'] + chicago['SCH_CNT'] + chicago['CC_CNT'] + chicago['SC_CNT'] + chicago['WC_CNT'] + chicago['SS_CNT']
```

```python
chicago = chicago.to_crs(3857)
chicago.to_file('data/chicago_tract_all.shp')
```

Population-weighted centroids

```python
cook = gpd.read_file('data/chicago_tract.shp')
cook_block = gpd.read_file('data/cook_block.shp')
pop = pd.read_csv('data/nhgis0008_ds258_2020_block.csv', encoding='latin-1')
cook_block = cook_block.merge(pop[['GISJOIN', 'U7H001']], on='GISJOIN')

# Calculate the population-weighted center for each tract in cook_tract
tract_centers = []
for _, tract in cook.iterrows():
    blocks_in_tract = cook_block[cook_block['geometry'].within(tract['geometry'])]
    
    weighted_center = blocks_in_tract['geometry'].centroid
    weighted_population = blocks_in_tract['U7H001']

    total_weighted_population = weighted_population.sum()

    if total_weighted_population > 0:
        # Use population-weighted centroid if population is greater than 0
        weighted_longitude = (weighted_center.x * weighted_population).sum() / total_weighted_population
        weighted_latitude = (weighted_center.y * weighted_population).sum() / total_weighted_population
    else:
        # Use geometric centroid if population is 0
        weighted_longitude = tract['geometry'].centroid.x
        weighted_latitude = tract['geometry'].centroid.y

    tract_centers.append({
        'GISJOIN': tract['GISJOIN'],
        'Lon_w': weighted_longitude,
        'Lat_w': weighted_latitude
    })

tract_centers_gdf = gpd.GeoDataFrame(tract_centers, geometry=gpd.points_from_xy([center['Lon_w'] for center in tract_centers], [center['Lat_w'] for center in tract_centers]))
cook_centroids = tract_centers_gdf.merge(cook, on='GISJOIN')
cook_centroids = cook_centroids.drop('geometry_y', axis=1)
cook_centroids = cook_centroids.rename(columns={'geometry_x': 'geometry'})
cook_centroids = cook_centroids.set_geometry('geometry')
cook_centroids.crs = cook.crs
cook_centroids.to_file('data/chicago_centroids_all.shp')
cook_centroids
```
